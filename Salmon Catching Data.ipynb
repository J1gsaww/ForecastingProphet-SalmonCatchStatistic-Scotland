#**Sum the coloumn and group by year**

```python
df['Year'] = df['Year'].astype(int) #ensure
columns_to_sum = [
      'Wild MSW number', 'Wild MSW weight (kg)',
      'Wild 1SW number', 'Wild 1SW weight (kg)',
      'Sea trout number', 'Sea trout weight (kg)',
      'Finnock number', 'Finnock weight (kg)',
      'Farmed MSW number', 'Farmed MSW weight (kg)',
      'Farmed 1SW number', 'Farmed 1SW weight (kg)'
      ]
yearly_summary = df.groupby('Year')[columns_to_sum].sum().reset_index()
print(yearly_summary)
```

#**Define Problem**


**Proportion of each type of fish caught each year**

```python
import matplotlib.pyplot as plt
import seaborn as sns
Wild_MSW = yearly_summary[['Year', 'Wild MSW number']].rename(columns={'Year': 'ds', 'Wild MSW number': 'y'})
Wild_1SW = yearly_summary[['Year', 'Wild 1SW number']].rename(columns={'Year': 'ds', 'Wild 1SW number': 'y'})
Sea_trout = yearly_summary[['Year', 'Sea trout number']].rename(columns={'Year': 'ds', 'Sea trout number': 'y'})


plt.figure(figsize=(18, 8))
sns.lineplot(x='ds', y='y', data=Wild_MSW, label='Wild MSW', color='red')
sns.lineplot(x='ds', y='y', data=Wild_1SW, label='Wild 1SW', color='orange')
sns.lineplot(x='ds', y='y', data=Sea_trout, label='Sea trout', color='yellow')
#sns.lineplot(x='ds', y='yhat', data=forecast, label='Predicted Wild MSW', color='orange')
plt.title('Proportion of each type of fish caught each year')
plt.xlabel('Year')
plt.ylabel('Fish Caught')
plt.legend()
plt.show()

filtered_yearly_summary = yearly_summary[(yearly_summary['Year'] >= 2012) & (yearly_summary['Year'] <= 2022)]
Wild_MSW = filtered_yearly_summary[['Year', 'Wild MSW number']].rename(columns={'Year': 'ds', 'Wild MSW number': 'y'})
Wild_1SW = filtered_yearly_summary[['Year', 'Wild 1SW number']].rename(columns={'Year': 'ds', 'Wild 1SW number': 'y'})
Sea_trout = filtered_yearly_summary[['Year', 'Sea trout number']].rename(columns={'Year': 'ds', 'Sea trout number': 'y'})
print('\n')

#2012 - 2022
plt.figure(figsize=(18, 8))
sns.lineplot(x='ds', y='y', data=Wild_MSW, label='Wild MSW', color='red')
sns.lineplot(x='ds', y='y', data=Wild_1SW, label='Wild 1SW', color='orange')
sns.lineplot(x='ds', y='y', data=Sea_trout, label='Sea trout', color='yellow')
plt.title('Proportion of each type of fish caught between 2012 - 2022')
plt.xlabel('Year')
plt.ylabel('Fish Caught')
plt.legend()
plt.show()
```

```python
filtered_yearly_summary = yearly_summary[(yearly_summary['Year'] >= 1994) & (yearly_summary['Year'] <= 2022)]
Farmed_MSW = filtered_yearly_summary[['Year', 'Farmed MSW number']].rename(columns={'Year': 'ds', 'Farmed MSW number': 'y'})
Farmed_1SW = filtered_yearly_summary[['Year', 'Farmed 1SW number']].rename(columns={'Year': 'ds', 'Farmed 1SW number': 'y'})

#1994 - 2022
plt.figure(figsize=(18, 8))
sns.lineplot(x='ds', y='y', data=Farmed_MSW, label='Farmed MSW', color='red')
sns.lineplot(x='ds', y='y', data=Farmed_1SW, label='Farmed 1SW', color='orange')
#sns.lineplot(x='ds', y='yhat', data=forecast, label='Predicted Wild MSW', color='orange')
plt.title('Proportion of each type of farmed salmon each year (1994 - 2022)')
plt.xlabel('Year')
plt.ylabel('Farmed Salmon')
plt.legend()
plt.show()


```

**Method of fishing**

1.Filter the data for the years 2017 to 2022.

2.Group the data by 'Year', 'Method', and sum the columns.

3.Plot the filtered data.

```python
unique_methods = df['Method'].unique()
print("Unique Methods of Fishing:")
print(unique_methods,'\n')

filtered_data = df[(df['Year'] >= 2017) & (df['Year'] <= 2022)]
columns_to_sum = filtered_data.select_dtypes(include=['number']).columns
group_columns = ['Year', 'Method']
columns_to_sum = [col for col in columns_to_sum if col not in group_columns]
method_summary = filtered_data.groupby(group_columns)[columns_to_sum].sum().reset_index()


print("Method Summary Head:")
print(method_summary.head())
```

```python
Wild_MSW = method_summary[['Year', 'Method', 'Wild MSW number']].rename(columns={'Year': 'ds', 'Wild MSW number': 'y'})
Wild_1SW = method_summary[['Year', 'Method', 'Wild 1SW number']].rename(columns={'Year': 'ds', 'Wild 1SW number': 'y'})
Sea_trout = method_summary[['Year', 'Method', 'Sea trout number']].rename(columns={'Year': 'ds', 'Sea trout number': 'y'})
Farmed_MSW = method_summary[['Year', 'Method', 'Farmed MSW number']].rename(columns={'Year': 'ds', 'Farmed MSW number': 'y'})
Farmed_1SW = method_summary[['Year', 'Method', 'Farmed 1SW number']].rename(columns={'Year': 'ds', 'Farmed 1SW number': 'y'})

plt.figure(figsize=(18, 8))
#palette is a set of color
sns.lineplot(x='ds', y='y', hue='Method', data=Wild_MSW, palette='Set1', legend='full')
#sns.lineplot(x='ds', y='y', hue='Method', data=Wild_1SW, palette='Set2', legend='full')
#sns.lineplot(x='ds', y='y', hue='Method', data=Sea_trout, palette='Set3', legend='full')
#sns.lineplot(x='ds', y='y', hue='Method', data=Farmed_MSW, palette='Set2', legend='full')
#sns.lineplot(x='ds', y='y', hue='Method', data=Farmed_1SW, palette='Set3', legend='full')

plt.title('Proportion of Wild MSW Caught by Method Each Year (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Fish Caught')
plt.legend(title='Method')
plt.show()
print('\n')

plt.figure(figsize=(18, 8))
sns.lineplot(x='ds', y='y', hue='Method', data=Wild_1SW, palette='Set1', legend='full')
plt.title('Proportion of Wild 1SW Caught by Method Each Year (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Fish Caught')
plt.legend(title='Method')
plt.show()
print('\n')

plt.figure(figsize=(18, 8))
sns.lineplot(x='ds', y='y', hue='Method', data=Sea_trout, palette='Set1', legend='full')
plt.title('Proportion of Sea trout Caught by Method Each Year (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Fish Caught')
plt.legend(title='Method')
plt.show()
print('\n')

plt.figure(figsize=(18, 8))
sns.lineplot(x='ds', y='y', hue='Method', data=Farmed_MSW, palette='Set1', legend='full')
plt.title('Proportion of Farmed MSW Caught by Method Each Year (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Fish Caught')
plt.legend(title='Method')
plt.show()
print('\n')

plt.figure(figsize=(18, 8))
sns.lineplot(x='ds', y='y', hue='Method', data=Farmed_1SW, palette='Set1', legend='full')
plt.title('Proportion of Farmed 1SW Caught by Method Each Year (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Fish Caught')
plt.legend(title='Method')
plt.show()
```

#Decompose

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose = seasonal_decompose(yearly_summary['Wild MSW number'], model='additive', extrapolate_trend='freq',period=5) #1 years
decompose.plot()
plt.show
```

**Trend: The long-term movement or general direction in the data"**

**Seasonality: fluctuations or patterns that repeat at regular intervals**

**Residual/error: Remainder or leftover variation in the data**

**Actual Time Series** = Trend + Seasonality + Residuals

#Forecasting

**yhat:** The predicted forecast values for each corresponding date (ds)

**yhat_lower:** The lower boundary of the predicted forecast, representing the lower uncertainty interval.

**yhat_upper:** The upper boundary of the predicted forecast, representing the upper uncertainty interval.

```python
#Train
prophet_data = yearly_summary[['Year', 'Wild MSW number']].rename(columns={'Year': 'ds', 'Wild MSW number': 'y'}) #y mean value to be forecasted column
prophet_data['ds'] = pd.to_datetime(prophet_data['ds'], format='%Y') #ds mean datestamp

#Fit the model
model = Prophet()
model.fit(prophet_data)

#Forecasting
future = model.make_future_dataframe(periods=5, freq='Y') #5 years
display(future)
#forecast = model.predict(future)
#freq = 'MS' --> month
#freq = 'Y' --> Year
forecast = model.predict(future)

#Ensure forecasted values are non-negative
#forecast['yhat'] = forecast['yhat'].clip(lower=0)  # Clip negative values to 0
display(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
#Plot the time series
fig = model.plot(forecast)
#Add a vertical line at the end of traning period
axes = fig.gca()
last_training = forecast['ds'].iloc[-5]
axes.axvline(last_training, color='red', linestyle='--',label='Traning End')
axes.legend()

plt.legend()
plt.title('Wild MSW Forecast')
plt.xlabel('Year')
plt.ylabel('Wild MSW Number')
plt.show()
```
